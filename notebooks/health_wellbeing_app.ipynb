### Generating synthetic data.

```python
import numpy as np
import pandas as pd
import uuid

#In the following functions we generate the data for the events dataframe.
def user_id(n):
    return [str(uuid.uuid4()) for _ in range(1, n + 1, 1)]

def datetime(n):
    return pd.date_range('2021-01-01', '2023-01-01', periods = n)

def session_length(n):
    return np.random.randint(1, 121, size = n)

def rating(n):
    return np.random.randint(1, 6, size = n)

def device(n):
    return np.random.choice(['iPhone', 'Android', 'Windows Phone', 'Desktop'], size = n) 

def status(n):
    return np.random.choice(['Complete', 'Incomplete'], size = n)


#Generate events dataframe.
def generate_df(n):
    return pd.DataFrame({
        'uid': user_id(n),
        'datetime': datetime(n),
        'session_length': session_length(n),
        'rating': rating(n),
        'device': device(n),
        'status': status(n)
    })
    
```

```python
df = generate_df(1000)
```
